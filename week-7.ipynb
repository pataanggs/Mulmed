```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numpy_img_3ch = np.array([
    [[0,120,0], [0,255,0], [255,0,0]],
    [[255,0,255], [0,255,255], [255,0,255]],
    [[192,192,192], [128,128,128], [64,64,64]],
], dtype=np.uint8)
```

```python
plt.imshow(numpy_img_3ch)
plt.axis('off')
plt.show()
```

```python
import cv2

#average all channels using numpy
gray_img = np.mean(numpy_img_3ch, axis=1).astype(np.uint8)

plt.imshow(gray_img, cmap='gray')
plt.axis('off')
plt.show()

print(numpy_img_3ch)
```

```python
import os
img = cv2.imread(os.path.join('Worksheet-2', 'data', 'IU-ByeSummer.jpg'))
plt.figure(figsize=(12,6))


plt.imshow(img)
plt.axis('off')
plt.title('Original Image')
plt.show()

#congvert to rgb
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#plot side by side
plt.figure(figsize=(12,6))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.axis('off')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(img_rgb)
plt.axis('off')
plt.title('RGB Image')

plt.show()
```

```python
# Tambah border hitam & merah lebih ringkas
import cv2, numpy as np, matplotlib.pyplot as plt

# Tambah border hitam 10px, lalu merah 5px
img_bordered = np.pad(img, ((15,15),(15,15),(0,0)), mode='constant', constant_values=0)

# Warnai border luar 5px jadi merah
img_bordered[:5] = [0,0,255]
img_bordered[-5:] = [0,0,255]
img_bordered[:, :5] = [0,0,255]
img_bordered[:, -5:] = [0,0,255]

# Tampilkan
plt.imshow(cv2.cvtColor(img_bordered, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

#print shape gambar
print('Shape gambar asli:', img.shape)
print('Shape gambar dengan border:', img_bordered.shape)
```

```python
img_cropped_original = img[100:400, 200:500]
img_cropped_bordered = img_bordered[115:415, 215:515]

# Crop gambar persis di tengah
h, w = img.shape[:2]
center_h, center_w = h // 2, w // 2
crop_size = 150  # setengah dari ukuran crop (300x300)

img_cropped_center = img[center_h-crop_size:center_h+crop_size, center_w-crop_size:center_w+crop_size]

plt.figure(figsize=(15,5))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img_cropped_original, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Cropped Original')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(img_cropped_bordered, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Cropped Bordered')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(img_cropped_center, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Cropped Center')

plt.show()
```
