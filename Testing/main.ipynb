```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Load audio files
audio1, sr1 = librosa.load('1.mp3')
audio2, sr2 = librosa.load('2.mp3')

# Create figure with subplots: 2 rows (one per file), 2 columns (waveform and spectrogram)
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# File 1: Waveform
axes[0, 0].plot(np.arange(len(audio1)) / sr1, audio1)
axes[0, 0].set_title('Waveform of 1.mp3', fontsize=14)
axes[0, 0].set_xlabel('Waktu (detik)')
axes[0, 0].set_ylabel('Amplitude')

# File 1: Spectrogram
D1 = librosa.stft(audio1)
S_db1 = librosa.amplitude_to_db(np.abs(D1), ref=np.max)
img1 = librosa.display.specshow(S_db1, sr=sr1, x_axis='time', y_axis='hz', ax=axes[0, 1], cmap='magma')
axes[0, 1].set_title('Spectrogram of 1.mp3 (Log-dB, STFT)', fontsize=14)
axes[0, 1].set_xlabel('Waktu (detik)')
axes[0, 1].set_ylabel('Frekuensi (Hz)')
axes[0, 1].set_ylim(20, 3000)  # Zoom in to 20Hz - 3kHz
fig.colorbar(img1, ax=axes[0, 1], format='%+2.0f dB')

# File 2: Waveform
axes[1, 0].plot(np.arange(len(audio2)) / sr2, audio2)
axes[1, 0].set_title('Waveform of 2.mp3', fontsize=14)
axes[1, 0].set_xlabel('Waktu (detik)')
axes[1, 0].set_ylabel('Amplitude')

# File 2: Spectrogram
D2 = librosa.stft(audio2)
S_db2 = librosa.amplitude_to_db(np.abs(D2), ref=np.max)
img2 = librosa.display.specshow(S_db2, sr=sr2, x_axis='time', y_axis='hz', ax=axes[1, 1], cmap='magma')
axes[1, 1].set_title('Spectrogram of 2.mp3 (Log-dB, STFT)', fontsize=14)
axes[1, 1].set_xlabel('Waktu (detik)')
axes[1, 1].set_ylabel('Frekuensi (Hz)')
axes[1, 1].set_ylim(20, 3000)  # Zoom in to 20Hz - 3kHz
fig.colorbar(img2, ax=axes[1, 1], format='%+2.0f dB')

plt.tight_layout()
plt.show()
```
